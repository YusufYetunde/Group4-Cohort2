```python
new_data=pd.DataFrame(data,(range(1,27820)),['country','year','sex','age','suicides_no','population','suicides/100k pop','gdp_per_capita ($)','generation'])
new_data.head()
```

```python
sex=pd.get_dummies(new_data['sex'],drop_first=True)
sex.head()
```

```python
generation=pd.get_dummies(new_data['generation'],drop_first=True)
generation.head()
```

```python
age=pd.get_dummies(new_data['age'],drop_first=True)
age.head()
```

```python
new_data.drop(['country','sex','age','generation'],axis=1,inplace=True)
new_data.head()
```

```python
df=pd.concat([new_data,sex,generation,age],axis=1)
df.head()
```

```python
y=df.iloc[:,1].values
```

```python
df_new=df.drop(['suicides_no'],axis=1)
```

```python
df_new.head()
```

```python
x=df_new.values
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0,test_size=0.25)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor= DecisionTreeRegressor(random_state=0)
```

```python
regressor.fit(x_train,y_train)
```

```python
y_pred=regressor.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,explained_variance_score
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
explained_variance_score(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
plt.scatter(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor= RandomForestRegressor(n_estimators=500,random_state=0)
```

```python
regressor.fit(x_train,y_train)
```

```python
y_pred=regressor.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,explained_variance_score
```

```python
mean_absolute_error(y_test,y_pred)
```
